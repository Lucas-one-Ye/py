（1）在同一张图中创建两个子图，分别画出sinx和cosx在[-3.14,3.14]上的函数图像。设置线条宽度为2.5.(提示：.linspace()，plt.subplot(),.plot(,linewidth=..)

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
#设置中文字体
plt.rcParams['font.sans-serif']=['SimHei']  #用来正常显示中文标签，设置为黑体
plt.rcParams['axes.unicode_minus']=False    #用来正常显示负号
import numpy as np
import matplotlib.pyplot as plt
#取x变量的数据
x=np.linspace(-3.14,3.14,50)

#获取sinx和cosx
y1=np.sin(x)
y2=np.cos(x)

#绘制第一个子图sinx
plt.subplot(121)
plt.plot(x,y1,linewidth=2.5)
#绘制第二个子图
plt.subplot(122)
plt.plot(x,y2,linewidth=2.5)
plt.show()
```

（8）绘制一个评分、评分人数和价格之间的相关系数图。（seaborn绘图）

```python
import pandas as pd
df2=df[['评分','评分人数','价格']]
df2
```

```python
#生成相关系数矩阵
a=np.corrcoef(df2,rowvar=0)
a
```

```python
 # rowvar=0表示每列为一个特征
a=pd.DataFrame(data=a,columns=df2.columns,index=df2.columns)
a
```

```python
##形式1
mask=np.zeros_like(a)  # 生成全0矩阵
mask
```

```python
mask[np.triu_indices_from(mask)]=True  #取上三角矩阵并设置为1
mask
```

```python
plt.figure(figsize=(8,8))
with sns.axes_style("white"): 
    ax=sns.heatmap(a,mask=mask,square=True,annot=True)#annot是否加标记
ax.set_title("评分、评分人数和价格之间的相关系数图",fontsize=18)
plt.show()
```

（10）用OpenCV实现对自己的图像进行人脸检测并做标记（可以是其他的五官；方法可以参考课堂介绍的示例，也可以查找资料利用其他方法完成）

```python
import numpy
import cv2
 
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
 
img = cv2.imread("meimei.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
 
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
 
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
